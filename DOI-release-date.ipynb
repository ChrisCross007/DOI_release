## DOI DATE SCRIPT (HBP)

**Description:** Script for fetching release dates from DOI:s using Datacite API.

**Author:** Krister Andersson (2019)


```python
import requests
import json
import pandas as pd
```

```python
results = []
with open('doi2.txt') as inputfile:
    for line in inputfile:
        ## line= line.strip("https://doi.org/")
        line = line.replace("\n",'')
        results.append(line)
        
print(results)
```

```python
#url = "https://api.datacite.org/dois?prefix:10.25493"
date = []
title = []
doi = []
for i in results:
    try:
        print(i)
        url = "https://api.datacite.org/works/" + i
        print(url)
        headers = {'accept': 'application/vnd.api+json'}
        response = requests.request("GET", url, headers=headers)
        data = response.json()
        di = data['data']['attributes']['registered']
        dt = data['data']['attributes']['title']
        date.append(di)
        doi.append(i)
        title.append(dt)
        
    except:
        print("An exception occurred")
        #date.append("N/A")
```

```python
data = zip(title,doi, date)
```

```python
df = pd.DataFrame(data, columns= ['Title', 'DOI','Date'])
```

```python
df.Date = pd.to_datetime(df.Date , format='%Y-%m-%d', errors='ignore')
df['Count'] = 1
```
